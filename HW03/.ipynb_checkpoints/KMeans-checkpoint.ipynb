## 2.建立迭代回圈  

在此迴圈中，一開始會先負責將所有資料點分類群集，最後再根據群集的結果更新中心點。


## 3.分類群集相關實作(caculate_Euclidean_distance)

一開始，會先將讀入的所有資料點以clustering_Euclidean，map成:  
(屬於第K簇中心點(與中心的距離,[[這個點的位置]]))  
接著用reduceByKey，將所有相同中心的資料加起來變成:  
(屬於第K簇中心點(所有點與中心的距離和,[屬於這個中心的所有點[point1],[point2]...]))  
中間為了拿出個別點的COST，簡單的用了
```=python

cost_list.append(finish_data.collect()[0][1])
```

## 4.更新中心點
前面我們已經將所有點分類好了，儲存於data_cent中  
這裡我用了map(recompute_cent)，將所有點相加，取平均，計算出這些點之中的新中心點  
最後用collectAsMap()把它直接轉回字典的形式，這樣可以達到更新字典的效果。  


## 5.分類群集相關實作(caculate_Manhattan_distance)
在曼哈頓距離中，所有時作細節與歐基里德距離相同，
只有替換掉計算距離的公式而已。

## 6.印出答案  


````python
from pyspark import SparkConf,SparkContext
from operator import add
import re

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

def caculate_Euclidean_distance(a,b):
    total=0
    for item in zip(a,b):
        total=total+((item[0]-item[1])** 2 )
    return(total**0.5)

def caculate_Manhattan_distance(a,b):
    total=0
    for item in zip(a,b):
        total=total+((item[0]-item[1]) )
    return(abs(total))

    
def clustering_Euclidean(points):
    global centroids
    global K
    mini_dist=float('inf')
    cent_num=0
    data=points.split(' ')
    data = list(map(float, data))
    for i in range(K):
        cent=centroids[i]
        cent=list(map(float, cent))
        euclDist=caculate_Euclidean_distance(data,cent)
        if(euclDist<mini_dist):
            mini_dist=euclDist
            cent_num=i
    return (cent_num,(mini_dist**2,[data]))

def clustering_Manhattan(points):
    global centroids
    global K
    mini_dist=float('inf')
    cent_num=0
    data=points.split(' ')
    data = list(map(float, data))
    for i in range(K):
        cent=centroids[i]
        cent=list(map(float, cent))
        ManhaDist=caculate_Manhattan_distance(data,cent)
        if(ManhaDist<mini_dist):
            mini_dist=ManhaDist
            cent_num=i
    return (cent_num,(mini_dist,[data]))


def list_add(a,b):
    c = []
    for i in range(len(a)):
        c.append(a[i]+b[i])
    return c

def recompute_cent(recompute_cent):
    global centroids
    origin_list=[0 for i in range(58)]
    length=len(recompute_cent[1][1])
    for points in recompute_cent[1][1]:
        origin_list=list_add(origin_list,points)
    centroids[recompute_cent[0]]=[os/length for os in origin_list]
    return (recompute_cent[0],centroids[recompute_cent[0]])

    
````

````python
sc.stop()
#本欄位為跑實際資料用的

K=10
MAX_ITER=20

cost_list=[]

conf=SparkConf().setMaster('local').setAppName('KMeans').set("spark.ui.port", "34050")
sc=SparkContext(conf=conf)
data = sc.textFile(r"./data.txt")
c1 = sc.textFile(r"./c1.txt")
c2 = sc.textFile(r"./c2.txt")

c1 = c1.map(lambda x: x.split(" ")).cache()
centroids =dict(zip(range(0,K),c1.collect()))
for i in range(MAX_ITER):
    #---群集---
    data_cent=data.map(clustering_Euclidean)
    data_cent=data_cent.reduceByKey(lambda x, y: (x[0]+y[0] ,x[1]+y[1]))
    finish_data=data_cent.map(lambda x:(0,x[1][0])).reduceByKey(lambda x, y:x+y)
    cost_list.append(finish_data.collect()[0][1])
    #---分群結束---
    #重新調整中心點
    centroids=data_cent.map(recompute_cent).collectAsMap()
c1_cost=cost_list
c1_center=centroids
````

````python
sc.stop()
#本欄位為跑實際c2資料用的

K=10
MAX_ITER=20

cost_list=[]

conf=SparkConf().setMaster('local').setAppName('KMeans').set("spark.ui.port", "34050")
sc=SparkContext(conf=conf)
data = sc.textFile(r"./data.txt")
c2 = sc.textFile(r"./c2.txt")

c2 = c2.map(lambda x: x.split(" ")).cache().collect()
centroids =dict(zip(range(0,K),c2))
for i in range(MAX_ITER):
    #---群集---
    data_cent=data.map(clustering_Euclidean)
    data_cent=data_cent.reduceByKey(lambda x, y: (x[0]+y[0] ,x[1]+y[1]))
    finish_data=data_cent.map(lambda x:(0,x[1][0])).reduceByKey(lambda x, y:x+y)
    cost_list.append(finish_data.collect()[0][1])
    #---分群結束---
    #重新調整中心點
    centroids=data_cent.map(recompute_cent).collectAsMap()

c2_cost=cost_list
c2_center=centroids
````

````python

bar=[str(i) for i in range(1,21)]
plt.figure(figsize=(10,4))
plt.plot(bar,c1_cost)
plt.plot(bar,c2_cost)
plt.ylabel('cost')
plt.xlabel('iter')

plt.show()
````

````python
import numpy as np
def caculate_percentage(cost_list):
    return abs((cost_list[-1]-cost_list[0])/(cost_list[0])*100)
def caculate_Euclidean_cent_dest(cent_dict,k):
    df_cent=pd.DataFrame()
    for i in range(k):
        cent_list=[]
        for j in range(k):
            if(i<j):
                cent_list.append(np.nan)
            else:
                cent_list.append(caculate_Euclidean_distance(cent_dict[i],cent_dict[j]))
        df_cent[i]=pd.Series(cent_list)
    return df_cent

def caculate_Manhattan_cent_dest(cent_dict,k):
    df_cent=pd.DataFrame()
    for i in range(k):
        cent_list=[]
        for j in range(k):
            if(i<j):
                cent_list.append(np.nan)
            else:
                cent_list.append(caculate_Manhattan_distance(cent_dict[i],cent_dict[j]))
        df_cent[i]=pd.Series(cent_list)
    return df_cent
````
